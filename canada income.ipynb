```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
print(df.head())
```

```python
plt.scatter(df['year'], df['per capita income (US$)'], marker='*', color='red')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Scatter Plot of Per Capita Income Over Years')
plt.show()
```

```python
X_linear = df[['year']]
y_linear = df['per capita income (US$)']
```

```python
X_train_linear, X_test_linear, y_train_linear, y_test_linear = train_test_split(X_linear, y_linear, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(X_train_linear, y_train_linear)
```

```python
linear_prediction_2020 = linear_model.predict([[2020]])
print("Linear Regression Prediction for 2020:", linear_prediction_2020[0])
```

```python
bins = [0, 15000, 30000, float('inf')]
labels = ['Low', 'Medium', 'High']
df['income_class'] = pd.cut(df['per capita income (US$)'], bins=bins, labels=labels)
```

```python
X_logistic = df[['year']]
y_logistic = df['income_class']
```

```python
X_train_logistic, X_test_logistic, y_train_logistic, y_test_logistic = train_test_split(X_logistic, y_logistic, test_size=0.2, random_state=42)
```

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train_logistic, y_train_logistic)
```

```python
y_pred_logistic = logistic_model.predict(X_test_logistic)
accuracy_logistic = accuracy_score(y_test_logistic, y_pred_logistic)
print("Logistic Regression Accuracy:", accuracy_logistic)
```

```python
X_tree = df[['year']]
y_tree = df['income_class']
```

```python
X_train_tree, X_test_tree, y_train_tree, y_test_tree = train_test_split(X_tree, y_tree, test_size=0.2, random_state=42)
```

```python
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train_tree, y_train_tree)
```

```python

y_pred_tree = tree_model.predict(X_test_tree)
accuracy_tree = accuracy_score(y_test_tree, y_pred_tree)
print("Decision Tree Classifier Accuracy:", accuracy_tree)
```

```python
models = ['Linear Regression', 'Logistic Regression', 'Decision Tree Classifier']
accuracies = [linear_model.score(X_linear, y_linear), accuracy_logistic, accuracy_tree]
```

```python
plt.bar(models, accuracies, color=['blue', 'green', 'orange'])
plt.ylabel('Accuracy Score')
plt.title('Comparison of Model Accuracy Scores')
plt.show()
```
